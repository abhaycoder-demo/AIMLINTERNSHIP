```python
import cv2
import numpy as np
image=cv2.imread('child.jpg')
cv2.imshow('image',image)
cv2.waitKey(5000)
model=cv2.CascadeClassifier(cv2.data.haarcascades+'haarcascade_frontalface_default.xml')
faces=model.detectMultiScale(image)
print(faces)
```

```python
import cv2
import numpy as np
image=cv2.imread('child.jpg')

model=cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
faces=model.detectMultiScale(image)
print(faces)
for face in faces:
    x,y,w,h=face
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,255,0),3)
cv2.imwrite('child.jpg',image)
```
