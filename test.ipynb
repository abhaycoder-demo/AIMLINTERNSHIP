```python
import numpy as np
from sklearn.linear_model import LinearRegression

# Features: Age, Height, BP
X = np.array([
    [20, 165, 110],
    [25, 170, 120],
    [30, 175, 125],
    [35, 180, 130],
    [40, 185, 140]
])

# Target: Weight
y = np.array([55, 62, 70, 78, 85])

# Create model
model = LinearRegression()

# Train model
model.fit(X, y)

# Predict weight for:
# Age = 28, Height = 172 cm, BP = 122
prediction = model.predict([[28, 172, 122]])

print("Predicted Weight:", prediction[0])
```
